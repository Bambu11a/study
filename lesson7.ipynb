```python
import math

class Money:
    def __init__(self,rubles:int,coins:int) -> None:
        self.__summa = rubles + coins / 100
        self.__rubles = self.__summa // 1
        self.__coins = (self.__summa%1) * 100  
        

    
    @property
    def rubles(self) ->int:
        return self.__rubles
    
    @rubles.setter
    def rubles(self, rubles: int):
        self.__rubles = rubles
    @property
    def coins(self) ->int:
        return self.__coins
    
    @coins.setter
    def coins(self, coins: int):
        self.__rubles += coins // 1
        self.__coins = coins % 1

    @property
    def summa(self) ->int:
        return self.__summa
    
    @summa.setter
    def summa(self, summa: int):
        self.__summa = summa
        self.__rubles = summa // 1
        self.__coins = summa % 1   

    def __add__(self, other):
        self.summa += other.summa
        self.coins = self.summa % 1
        self.rubles = self.summa // 1 * 100
        other.rubles, other.coins =0, 0 
        return self.summa

    def __sub__(self, other):
        self.summa -= other.summa
        self.coins = self.summa % 1
        self.rubles = self.summa // 1 * 100
        other.rubles, other.coins =0, 0 
        return 

    def __mul__(self, other: float):

        self.summa = math.ceil((self.rubles * 100 + self.coins) * other)
    
        self.rubles, self.coins= int(self.summa  // 1), (self.summa  % 1 * 100)

        return self.summa 

    def __truediv__(self, other):

        if isinstance(other, Money):

            value = (self.rubles*100 + self.coins) / (other.rubles*100 + other.coins)

            return f" differs {value}"

        else:
            value = math.ceil((self.rubles * 100 + self.coins) / other)
            self.rubles, self.coins = value // 100, value % 100

            return value

    def __eq__(self, other):

        return (self.summa) == (other.summa)

    def __ne__(self, other):

        return (self.summa) != (other.summa)

    def __lt__(self, other):

        return (self.summa) < (other.summa)

    def __gt__(self, other):

        return (self.summa) > (other.summa)

    def __le__(self, other):

        return (self.summa) <= (other.summa)

    def __ge__(self, other):

        return (self.summa) >= (other.summa)
```
