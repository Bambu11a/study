```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
from sklearn import preprocessing
```

```python
df = pd.read_csv("timesData.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
country = list(df["country"].unique())
country
```

```python
student_staff_rate = [] 
for i in country:
    x = df[df["country"] == i] 
    
    student_ratio = sum(x.student_staff_ratio) / len(x)
    student_staff_rate.append(student_ratio) 
    
dataframe = pd.DataFrame({"country":country , "student_staff":student_staff_rate }) 


plt.figure(figsize = (17,12))
sns.barplot(x = dataframe["country"] , y = dataframe["student_staff"] )
plt.xticks(rotation = 90)
plt.xlabel("Student_staf")
plt.ylabel("Country")
plt.show()

```

```python
teaching_rate = [] 
for i in country:
    x = df[df["country"] == i] 
    
    teaching_ratio = sum(x.teaching) / len(x)
    teaching_rate.append(teaching_ratio) 
    
dataframe = pd.DataFrame({"country":country , "teaching":teaching_rate }) 


plt.figure(figsize = (17,12))
sns.barplot(x = dataframe["country"] , y = dataframe["teaching"] )
plt.xticks(rotation = 90)
plt.xlabel("Country")
plt.ylabel("Teaching")
plt.show()
```

```python
df = pd.read_csv("Dataset salary.csv")
df.head()
```

```python
job_title = list(df["job_title"].unique())
len(job_title)
```

```python
company_size = list(df["company_size"].unique())
company_size
```

```python
salary_in_usd_rate = [] 
for i in job_title:
    x = df[df["job_title"] == i] 
    
    salary_in_usd_ratio = sum(x.salary_in_usd) / len(x)
    salary_in_usd_rate.append(salary_in_usd_ratio) 
    
dataframe = pd.DataFrame({"job_title":job_title , "salary_in_usd":salary_in_usd_rate }) 


plt.figure(figsize = (30,15))
sns.barplot(x = dataframe["job_title"] , y = dataframe["salary_in_usd"] )
plt.xticks(rotation = 90)
plt.xlabel("Job")
plt.ylabel("Salary")
plt.show()
```
