Стоимость биллетов привязанная к классам и портам посадки

```python
data_fare = data_train.groupby(["Pclass", "Embarked"]).agg({"Fare": ["min", "mean", "max", "count"]})
data_fare.columns = ["min", "mean", "max", "count"]
data_fare.reset_index(inplace=True)
data_fare
```

```python
def get_last_name(name:str) -> str:
    return name.split(",")[0]
```

Родственники

```python
data_train["Last name"] = data_train["Name"].apply(get_last_name)
data_train
```

```python
data_sibsp = data_train[data_train["SibSp"]>=1]
data_sibsp = data_sibsp.sort_values(by=['Last name'])
data_sibsp.head(20)
```

```python
data_parch = data_train[data_train["Parch"]>=1]
data_parch = data_parch.sort_values(by=['Last name'])
data_parch.head(20)
```

```python
data_surv = data_train[data_train.Survived==1]
data_surv
```

Сколько мужчин и женщин выжило

```python
data_surv.groupby("Sex")["PassengerId"].count().reset_index(name = "Count")
```

Сколько выживших в зависимости от класса

```python
data_surv.groupby(["Sex", "Pclass", "Embarked",])["Pclass"].count().reset_index(name = "Count")
```
